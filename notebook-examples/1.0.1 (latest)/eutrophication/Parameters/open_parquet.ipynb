```python
import pandas as pd
import ipywidgets as widgets
import os

# List all parquet files in the beacon_output folder
files = [f for f in os.listdir('beacon_output') if f.endswith('.parquet')]

# Create a dropdown widget for file selection
file_selector = widgets.Dropdown(
    options=files,
    description='Select file:',
    disabled=False,
)

display(file_selector)

# Load the selected file into a DataFrame
def load_file(change):
    df = pd.read_parquet(f'beacon_output/{change.new}')
    display(df.head())

file_selector.observe(load_file, names='value')
```

```python
print("selected file: ", file_selector.value)
```

```python
# Display the currently selected dataframe
df = pd.read_parquet(f'beacon_output/{file_selector.value}')
display(df.head())
```
